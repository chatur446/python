```python
# Array
A = [1,2,3]
prin[ ]
t(A)
```

```python
# Append - Insert element at end of array - on average:o(1)
A.append(5)
print(A)
```

```python
# pop - Deleting element at end of array - o(1)
A.pop()
print(A)
```

```python
# Insert ( not at end of array) - o(n)
A.insert(2,5)
print(A)
```

```python
# Modify an element - o(1)
A[0]=7
print(A)
```

```python
# Accessing element given index - o(1)
print(A[2])
```

```python
# Checking if array has an element - o(n)
if 7 in A:
  print(True)
```

```python
# Checking length - o(1)
print(len(A))
```

```python
# String
# Append to end of string - o(n)
s = 'hello'
b = s + 'z'
print(b)
```

```python
# Check if something is in string - o(n)
if 'h' in s:
  print(True)
```
